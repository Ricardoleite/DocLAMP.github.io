## Constantes de Ajuste

Com base nos resultados numéricos as constantes $C_{0}$, $C_{1}$ e $C_{2}$ são ajustadas para o caso de 1 bloco. Dado a correlação obtida

$$
\overline{Nu}= \displaystyle\frac{L}{H} \gamma^{-1/2} =\displaystyle\frac{L}{H} \left( C_{1} \left(Ra_{H}\left(\displaystyle\frac{Pr}{C_{0}+Pr}\right)\right)^{-1/2} -C_{2}  Ha^{2} Ra_{H}^{-1} \right)^{-1/2},
$$

as constantes $C_{0}$ e $C_{1}$ são ajustados com base nos resultados obtidos $Ha=0$ e a constante $C_{2}$ como uma média dos valores ajustados para $Ha=50$ e $Ha=100$;



 ### Ajuste de $C_{0}$ e $C_{1}$ para Ha=0
 
 Com base no ajuste feito foram obtidos $C_{0}=0,112888$ e $C_{1}=0,00649$. A figura abaixo ilustra o ajuste numérico.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
N1=np.array([3.59,3.88,4.21,4.27,6.95,7.44,8.5,8.74])
gamma=np.ones((len(Ra),len(Pr)))
# A1=0.004496
# B1=0.0051449
A1=0.00649
B1=0.112888
for j in range (len(Ra)):
    for i in range (len(Pr)):
        gamma[j][i]=(A1*Ra[j]*Pr[i]/(B1+Pr[i]))**(-1./2.)
gama=gamma.reshape(len(Pr)*len(Ra))

plt.semilogx(gama,N1,'ro',fillstyle='none')
CN1=gama
plt.semilogx(gama,CN1**(-1./2.),'r--',fillstyle='none')
plt.ylabel('$Nu_{av}$', color='k', horizontalalignment='right',rotation=0, fontsize=20)
plt.xlabel('$\\gamma$', fontsize=20)
plt.legend(['LBM','Correlação'],fontsize=11.,bbox_to_anchor=(1.0, 1.02))
plt.show()
```

 ### Correlação Ajustada
 
 Tirando a média de $C_{2}$, tem-se $C_{2}$=-5,5. Portanto, a correlação ajustada é dada por
 
 $$
\overline{Nu}= \displaystyle\frac{L}{H} \gamma^{-1/2} =\displaystyle\frac{L}{H} \left( 0,00649 \left(Ra_{H}\left(\displaystyle\frac{Pr}{0,112888
  +Pr}\right)\right)^{-1/2} +5,5  Ha^{2} Ra_{H}^{-1} \right)^{-1/2}.
$$
 
 O gráfico abaixo ilustra o ajuste geral da correlação obtida.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
N1H0=np.array([3.59,3.88,4.21,4.27,6.95,7.44,8.5,8.74])
N1H50=np.array([1.89,1.90,1.91,1.91,5.64,5.87,6.18,6.22])
N1H100=np.array([1.16,1.16,1.16,1.16,3.64,3.68,3.71,3.71])
Ha=np.array([0.0,50.0,100.0])
Pr=np.array([0.050,0.100,0.500,1.000])
Ra=np.array([10.0**5,10.0**6])
gamma=np.ones((len(Ra),len(Pr),len(Ha)))
gammaH0=np.ones((len(Ra),len(Pr)))
gammaH50=np.ones((len(Ra),len(Pr)))
gammaH100=np.ones((len(Ra),len(Pr)))
A1=0.00649
B1=0.112888
C1=-5.5
for j in range (len(Ra)):
    for i in range (len(Pr)):
        gammaH0[j][i]=(A1*Ra[j]*Pr[i]/(B1+Pr[i]))**(-1./2.)-C1*(Ha[0]**2.)/Ra[j]
        gammaH50[j][i]=(A1*Ra[j]*Pr[i]/(B1+Pr[i]))**(-1./2.)-C1*(Ha[1]**2.)/Ra[j]
        gammaH100[j][i]=(A1*Ra[j]*Pr[i]/(B1+Pr[i]))**(-1./2.)-C1*(Ha[2]**2.)/Ra[j]
        for k in range (len(Ha)):
            gamma[j][i][k]=(A1*Ra[j]*Pr[i]/(B1+Pr[i]))**(-1./2.)-C1*(Ha[k]**2.)/Ra[j]
gama=gamma.reshape(len(Ra)*len(Pr)*len(Ha))
gamaH0=gammaH0.reshape(len(Ra)*len(Pr))
gamaH50=gammaH50.reshape(len(Ra)*len(Pr))
gamaH100=gammaH100.reshape(len(Ra)*len(Pr))

#plt.semilogx(gama,N1H100,'ro:',fillstyle='none')
CN1=gama**(-1./2.)
CNH0=gamaH0**(-1./2.)
CNH50=gamaH50**(-1./2.)
CNH100=gamaH100**(-1./2.)
plt.semilogx(gamaH0,CNH0,'r:',fillstyle='none')
plt.semilogx(gamaH0,N1H0,'rs',fillstyle='none')
plt.semilogx(gamaH50,CNH50,'b:',fillstyle='none')
plt.semilogx(gamaH50,N1H50,'bv',fillstyle='none')
plt.semilogx(gamaH100,CNH100,'k:',fillstyle='none')
plt.semilogx(gamaH100,N1H100,'k>',fillstyle='none')
plt.ylabel('$Nu_{av}$', color='k', horizontalalignment='right',rotation=0, fontsize=20)
plt.xlabel('$\\gamma$', fontsize=20)
plt.legend(['LBM-Ha=0','Correlação-Ha=0','LBM-Ha=50','Correlação-Ha=50','LBM-Ha=100','Correlação-Ha=100'],fontsize=11.,bbox_to_anchor=(1.0, 1.02))
plt.show()
```
